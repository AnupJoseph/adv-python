```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.callbacks import TensorBoard
from keras.models import Sequential
from keras.layers import Dense
```

```python
optimizers = ["adam","RMSProp"]
initializers = ["GlorotUniform","normal"]
```

```python
for optimizer in optimizers:
    for initializer in initializers:
        model_name = f"{initializer} & {optimizers}"
        X,y = make_regression(n_samples=1000,n_features=6,random_state=0,noise=3.5,bias=0.4)
        X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

        model = Sequential()
        model.add(Dense(units = 6,kernel_initializer = initializer,activation = 'relu',input_dim = 6))
        model.add(Dense(units = 6,kernel_initializer = initializer,activation = 'relu'))
        model.add(Dense(units = 1,kernel_initializer = initializer))

        model.compile(loss = 'mean_squared_error',optimizer = optimizer,metrics = ['mean_absolute_percentage_error'])
        Board = TensorBoard(log_dir = 'logs/{}'.format(model_name))
        model_history = model.fit(X_train,y_train,epochs=100,callbacks = [Board])
```
