```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.rcParams["figure.figsize"] = (15,10)
```

```
bird_data = sns.load_dataset('penguins')
grid = sns.FacetGrid(bird_data,col='island',hue='species')
grid.map(plt.scatter,"flipper_length_mm", "body_mass_g",alpha=0.6)
grid.add_legend()
plt.show()
```

```
# Now lets try and get the weight of penguins organized by species for each island
sns.stripplot(x='island',y='body_mass_g',hue='species',data=bird_data,palette='Set2')
plt.show()
```

```
# In the above example its difficult to get inforamtion about the dream island
sns.swarmplot(x='island',y='body_mass_g',hue='species',data=bird_data,palette='Spectral_r',dodge=True)
plt.show()
```

```
sns.lmplot(x='body_mass_g',y='culmen_length_mm',hue='sex',col='island',markers=['o','x'],palette='Set2',data=bird_data)
plt.show()
```

```
sns.lmplot(x="body_mass_g", y="culmen_length_mm", hue="sex", col="island",row="sex",order=2, markers=["o", "x"],palette="Set1",data=bird_data)
plt.show()
```

```
sns.set_palette("gist_rainbow_r")
sns.jointplot(x="body_mass_g", y="culmen_length_mm", kind="hex",data=bird_data )
plt.show()
```

```
g = sns.jointplot(x="body_mass_g", y="culmen_length_mm", data=bird_data, kind="hex", color="c")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.set_axis_labels("Body Mass (in gram)", "Culmen Length ( in mm)")
plt.show()
```

```
sns.pairplot(bird_data, hue="island") 
plt.show()
```
